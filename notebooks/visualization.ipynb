```python
# import the libraries
import pandas as pd
from nltk.tokenize import word_tokenize
from pprint import pprint
import nltk
from nltk.draw.dispersion import dispersion_plot
import matplotlib.pyplot as plt

# make matplotlib inline
%matplotlib inline
```

```python
# load the data frame
df = pd.read_csv("../output/preprocessed_tweets_with_userid.csv", names=['userid', 'tweets'], delimiter='|')
pprint(df.head())
```

```python
# Removing non-ASCII characters in tweets and making it as s string
df['tweets'] = df.tweets.str.replace(r'\W+',' ')
pprint(df.head())
```

```python
# Creating a nltk text for the 1st userid in data frame: userid: 932563433605812226
text = nltk.Text(word_tokenize(df['tweets'][0]))
pprint(text)
```

```python
# Drawing the dispersion plotin reverse chronological order (recent tweets - past tweets)
plt.figure(figsize=(15, 5))
topics = ['follow', 'chance', 'india', 'prize']
dispersion_plot(text, topics)
```

```python
# Finding the frequency distribution for the top 50 common words in the text
words = word_tokenize(df['tweets'][0])
fdist = nltk.FreqDist(words)
plt.figure(figsize=(15, 5))
fdist.plot(50,cumulative=False)
```
